##  Bubble chart: Difficulty vs Popularity

```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Angular Why": {"popularity": 104.58, "difficulty":14.15},
    "Angular What": {"popularity": 201.60, "difficulty": 20.79},
    "Angular How": {"popularity": 143.95, "difficulty": 47.78},
    "Angular Other": {"popularity": 235.34, "difficulty": 17.25},
    "Webpack Why": {"popularity": 790.48, "difficulty":13.89},
    "Webpack What": {"popularity": 299.41, "difficulty": 22.03},
    "Webpack How ": {"popularity": 396.80, "difficulty": 47.20},
    "Webpack Other": {"popularity": 126.77, "difficulty": 16.98},
    "Reactjs Why": {"popularity": 184.92, "difficulty":20.312},
    "Reactjs What": {"popularity": 121.12, "difficulty": 19.53},
    "Reactjs How": {"popularity": 108.00, "difficulty": 45.70},
    "Reactjs Other": {"popularity": 201.64, "difficulty": 14.45},
    "React-native Why ": {"popularity": 149.56, "difficulty":16.94},
    "React-native What": {"popularity": 253.20, "difficulty": 20},
    "React-native How": {"popularity": 209.69, "difficulty": 46.44},
    "React-native Other": {"popularity": 103.66, "difficulty": 16.6},
    "Typescript Why": {"popularity": 419.79, "difficulty":17.76},
    "Typescript What": {"popularity": 535.69, "difficulty": 20.24},
    "Typescript How": {"popularity": 330.66, "difficulty": 53.71},
    "Typescript Other": {"popularity": 231.55, "difficulty": 8.26},
}


df = pd.DataFrame(data).T

plt.figure(figsize=(12,12))
scatter=plt.scatter(df['difficulty'], df['popularity'], s=df['popularity'] * 10, alpha=0.7,c=df['popularity'] * 10)
plt.rcParams['font.family'] = "Times New Roman"
plt.rcParams['font.size'] = 10
plt.colorbar(scatter)
for i, row in df.iterrows():
    plt.annotate(i, (row['difficulty'], row['popularity']), textcoords="offset points", xytext=(10,10), ha='center')


plt.grid(True)

plt.xlabel("Difficulty (% w/o Accepted Answers)")
plt.ylabel("Popularity (Avg. Views)")
plt.title("Bubble Chart: Difficulty vs. Popularity")

plt.show()
```

###  Library Treanding Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

final_1930['QCreationDate'] = pd.to_datetime(final_1930['QCreationDate'])

final_1930['YearMonth'] = final_1930['QCreationDate'].dt.to_period('Y')

final_1930['YearMonthStr'] = final_1930['YearMonth'].astype(str)

grouped = final_1930.groupby(['YearMonthStr', 'Library']).size().reset_index(name='QId')

plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped, x='YearMonthStr', y='QId', hue='Library', marker='o')

plt.title('Number of Posts Over Time by Library', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Number of Posts', fontsize=12)
plt.xticks(rotation=45)  
plt.grid(True)
plt.rcParams['font.family'] = "Times New Roman"
plt.rcParams['font.size'] = 10

plt.tight_layout()
plt.show()

```
